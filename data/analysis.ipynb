```python
import pandas as pd
import gffutils
```

```python
df_shi = pd.read_csv('shi_significant.csv')
df_shi.rename(columns={'Unnamed: 0': 'gene'}, inplace=True)
df_shi
```

```python
df_eco = pd.read_csv('eco_significant.csv')
df_eco.rename(columns={'Unnamed: 0': 'gene'}, inplace=True)
df_eco
```

```python
# sort based on absolute log2FoldChange
df_shi = df_shi.reindex(df_shi['log2FoldChange'].abs().sort_values(ascending=False).index)
df_eco = df_eco.reindex(df_eco['log2FoldChange'].abs().sort_values(ascending=False).index)
```

```python
df_eco
```

```python
df_shi
```

```python
# load orthologs
df_orthologs = pd.read_csv('orthologs.txt', sep='\t', header=None)

# remove duplicate values
df_orthologs = df_orthologs[df_orthologs[0].str.startswith('eco_protein')]

# split the orthologs
df_orthologs[0] = df_orthologs[0].str.split('|').str[1]
df_orthologs[1] = df_orthologs[1].str.split('|').str[1]

df_orthologs
```

```python
# replace cds- prefix
df_eco['gene'] = df_eco['gene'].str.replace('cds-', '')
df_shi['gene'] = df_shi['gene'].str.replace('cds-', '')
```

```python
# merge the orthologs with the significant genes
df_orthologs = pd.merge(df_eco, df_orthologs, left_on='gene', right_on=0)
df_orthologs = pd.merge(df_shi, df_orthologs, left_on='gene', right_on=1)

df_orthologs = df_orthologs.reindex(df_orthologs[2].abs().sort_values(ascending=False).index)
```

```python
COLUMN_MAP = {0: 'gene_eco_cds', 1: 'gene_shi_cds', 'log2FoldChange_x' : 'log2_eco', 'log2FoldChange_y': 'log2_shi'}

df_orthologs.rename(columns=COLUMN_MAP, inplace=True)
to_drop = [column for column in df_orthologs.columns if column not in COLUMN_MAP.values()]

df_orthologs.drop(columns=to_drop, inplace=True)

# sort columns by name
df_orthologs = df_orthologs.reindex(sorted(df_orthologs.columns), axis=1)
df_orthologs
```

```python
eco_db = gffutils.create.create_db('ref/eco_gff.gff', 'eco.db', merge_strategy='create_unique')
shi_db = gffutils.create.create_db('ref/shizo_gff.gff', 'shi.db', merge_strategy='create_unique')
```

```python
from collections import defaultdict

def dict_factory(row):
    d = defaultdict(lambda x: [''])
    for key in row.keys():
        d[key] = row[key]
    return d

def get_gene_name(db, gene_id):
    gene = db.execute(f'SELECT * FROM features WHERE id  = \'cds-{gene_id}\'').fetchone()
    if gene is None:
        return None
    
    data = defaultdict(lambda: '')

    data.update(eval(dict_factory(gene)['attributes']))
    
    return data['gene'], data['product']
```

```python
# add gene names to the dataframe
df_orthologs['gene_eco'] = df_orthologs['gene_eco_cds'].apply(lambda x: get_gene_name(eco_db, x)[0])
df_orthologs['gene_eco_prod'] = df_orthologs['gene_eco_cds'].apply(lambda x: get_gene_name(eco_db, x)[1])

df_orthologs['gene_shi'] = df_orthologs['gene_shi_cds'].apply(lambda x: get_gene_name(shi_db, x)[0])
df_orthologs['gene_shi_prod'] = df_orthologs['gene_shi_cds'].apply(lambda x: get_gene_name(shi_db, x)[1])

df_orthologs
```

```python
# generate plot for the orthologs
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')
plt.figure(figsize=(10, 10))
sns.scatterplot(data=df_orthologs, x='log2_eco', y='log2_shi')
plt.xlabel('log2FoldChange E. coli')
plt.ylabel('log2FoldChange S. Plomb')
plt.title('Orthologs log2FoldChange comparison')
plt.savefig('orthologs.png')
for i, row in df_orthologs.iterrows():
    plt.text(row['log2_eco'] + 0.01, row['log2_shi'] + 0.01, f"{row['gene_eco']} - {row['gene_shi']}", fontsize=8)
plt.show()

# BAB38639.1
# CAB91169.1
```
